```python
import numpy as np
import pandas as pd

import requests
from zipfile import ZipFile
from io import BytesIO
```

```python
txt_raw = 'cup98LRN.txt'
zip_raw = 'https://archive.ics.uci.edu/ml/machine-learning-databases/kddcup98-mld/epsilon_mirror/cup98lrn.zip'

r = requests.get(zip_raw)
files = ZipFile(BytesIO(r.content))
df = pd.read_csv(files.open(txt_raw), sep=",")
df
```

```python
df.info(verbose=True)
```

```python
pd.set_option('display.max_rows', None)
print(df.isna().sum())
pd.reset_option("display")
```

```python
def rmissingvaluecol(dff, threshold):
    l = []
    l = list(dff.drop(dff.loc[:,list((100*(dff.isnull().sum()/len(dff.index)) >= threshold))].columns, 1).columns.values)
    print("# Columns having more than %s percent missing values: "%threshold, (dff.shape[1] - len(l)))
    print("Columns:\n", list(set(list((dff.columns.values))) - set(l)))
    return l


rmissingvaluecol(df, 25)
```

```python
limitPer = len(df)*.25
df = df[df.isnull().sum(axis=1) < limitPer]
df = df.dropna(thresh=limitPer, axis=1)
df
```

```python
pd.set_option('display.max_rows', None)
print(df.isna().sum())
pd.reset_option("display")
```

```python
# Delete rows containing either 25% or more than 25% NaN Values
perc = 25.0

min_count =  int(((100-perc)/100)*df.shape[1] + 1)

mod_df = df.dropna(thresh=min_count)
mod_df = df.dropna(thresh=2)
mod_df
```

```python
print("Number of rows with at least 1 NA value: ", (len(df)-len(mod_df)))
```
